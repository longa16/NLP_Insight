```python
import pandas as pd
import numpy as np
import re

df  = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')

df = df[['Review Text', 'Rating', 'Class Name', 'Title']]
df = df.dropna(subset=['Review Text'])

def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text

df['Cleaned_Text'] = df['Review Text'].apply(clean_text)

print(f"Nombre d'avis chargés : {len(df)}")
print("--- Exemple avant/après nettoyage ---")
print(f"Original : {df['Review Text'].iloc[0]}")
print(f"Nettoyé  : {df['Cleaned_Text'].iloc[0]}")
```

```python
#topic modeling
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.decomposition import NMF

tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english') 
dtm = tfidf_vectorizer.fit_transform(df['Cleaned_Text'])
print(f"Forme de la matrice : {dtm.shape}")

nmf_model = NMF(n_components=5, random_state=42)
nmf_model.fit(dtm)

print("Modèle entraîné avec succès !")
```

```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"---Sujet {topic_idx + 1} ---")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]) )
        print("\n")

no_top_words = 15
display_topics(nmf_model, tfidf_vectorizer.get_feature_names_out(), no_top_words)
```

```python
# 1. On calcule la probabilité d'appartenance de chaque avis à chaque sujet
topic_values = nmf_model.transform(dtm)

df['Topic'] = topic_values.argmax(axis=1)

# 3. Affichage d'un exemple concret pour chaque sujet
for i in range(5):
    print(f"\n--- EXEMPLE DU SUJET {i} ---")
    topic_words = [tfidf_vectorizer.get_feature_names_out()[index] for index in nmf_model.components_[i].argsort()[:-11:-1]]
    print(f"Mots clés : {topic_words}")
    
    top_review_idx = topic_values[:, i].argmax()
    review_content = df.iloc[top_review_idx]['Review Text']
    
    print(f"Avis Représentatif :\n{review_content[:300]}...") # On affiche les 300 premiers caractères
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. On calcule la note moyenne pour chaque sujet
topic_stats = df.groupby('Topic')['Rating'].mean().sort_values()

topic_counts = df['Topic'].value_counts()

print("Note Moyenne par Sujet (du pire au meilleur) :")
print(topic_stats)

# 3. Visualisation
plt.figure(figsize=(10, 6))
colors = ['red' if x == topic_stats.min() else 'grey' for x in topic_stats.values] 

sns.barplot(x=topic_stats.index, y=topic_stats.values, palette=colors)
plt.title('Satisfaction Client par Sujet de Discussion', fontsize=15)
plt.xlabel('Numéro du Sujet', fontsize=12)
plt.ylabel('Note Moyenne', fontsize=12)
plt.ylim(3, 5) 
plt.axhline(df['Rating'].mean(), color='black', linestyle='--', label='Moyenne Globale')
plt.legend()
plt.show()
```
